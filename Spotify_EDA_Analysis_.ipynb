```python
# which artist has more  and less no of songs
song_list=df.groupby('artists')['name'].count()
song_df=song_list.reset_index(name='count')
song_data=song_df.sort_values(by='count')
highest_song=song_data.tail(10)
plt.figure(figsize=(12,8))
colors=plt.cm.viridis(highest_song['count']/highest_song['count'].max())
explode=[0] * len(highest_song)
explode[-1]=0.1
plt.pie(
    highest_song['count'],
    labels=highest_song['artists'],
    autopct='%1.1f',
    startangle=140,
    colors=colors,
    shadow=True,
    explode=explode
)
plt.title('highest songs per artist ')
plt.show()
```

```python
# which album has more and less  no of songs
mns=df.groupby('album')['name'].count()
mns_df=mns.reset_index(name='count')
mns_data=mns_df.sort_values(by='count')
highest_alb=mns_data.tail(10)
plt.figure(figsize=(10, 8))
colors = plt.cm.viridis(highest_alb['count'] / highest_alb['count'].max())
bars = plt.barh(

    highest_alb['album'],
    highest_alb['count'],
    color=colors,
)

for bar in bars:
    plt.text(bar.get_width(), bar.get_y() + bar.get_height()/2, f'{int(bar.get_width())}', va='center')

# Add labels and title
plt.xlabel('Number of Songs')
plt.ylabel('Albums')
plt.title(' Highest Number of Songs in Each Album')

# Display the horizontal bar chart
plt.show()
```

```python
# for each artist what is their loudness and dacek
artist_dance=df.groupby('artists')[['danceability', 'energy',
        'loudness',  'speechiness',
       'instrumentalness', 'liveness',  'tempo']].mean()
artist_df=artist_dance.sort_values( by= ['danceability', 'energy',
        'loudness',  'speechiness',
       'instrumentalness', 'liveness',  'tempo'])
l_artist=artist_df.head(10)
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have 'artists' as an index in your DataFrame
fig, axes = plt.subplots(5, 5, figsize=(30, 30))
fig.subplots_adjust(hspace=0.8, wspace=0.8)

# Flatten the axes for easy iteration
axes = axes.flatten()

# Define a color palette for pie charts
colors = plt.cm.viridis(range(len(l_artist.columns)))

# Loop through each artist and create a pie chart in each subplot
for i, (artist, row) in enumerate(l_artist.iterrows()):
    ax = axes[i]
    ax.pie(row.abs(), labels=row.index, autopct='%1.1f%%', startangle=180, colors=colors)
    ax.set_title(artist, fontsize=10)
    ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Hide any remaining empty subplots
for i in range(len(l_artist), len(axes)):
    fig.delaxes(axes[i])

# Adjust layout to prevent overlapping titles
plt.tight_layout()

# Overall title for the entire figure
plt.suptitle('Distribution of Audio Features for Each Artist', y=1.02, fontsize=16)

plt.show()



```

```python
# for each year how much duration is their
year_dur=df.groupby('year')['duration_ms'].mean()
year_dur_df = year_dur.reset_index(name='mean_duration_ms')
sorted_years_df = year_dur_df.sort_values(by='year')

highest_duration=sorted_years_df.tail(10)
plt.figure(figsize=(10, 6))
plt.fill_between(highest_duration['year'], highest_duration['mean_duration_ms'], color='skyblue', alpha=0.4)
plt.plot(highest_duration['year'], highest_duration['mean_duration_ms'], marker='o', color='blue')
plt.xlabel('Year')
plt.ylabel('Mean Duration (ms)')
plt.title('Mean Duration of Songs Over Each Year (Area Plot)')
plt.grid(True)
plt.show()
```

```python
lowest_duration=sorted_years_df.head(10)
plt.figure(figsize=(10, 6))
plt.fill_between(lowest_duration['year'], lowest_duration['mean_duration_ms'], color='skyblue', alpha=0.4)
plt.plot(lowest_duration['year'], lowest_duration['mean_duration_ms'], marker='o', color='blue')
plt.xlabel('Year')
plt.ylabel('Mean Duration (ms)')
plt.title('Mean Duration of Songs Over Each Year (Area Plot)')
plt.grid(True)
plt.show()
```

```python
# year and what is the range of the danceabiltiy and eneregy
year_d=df.groupby('year')[['danceability', 'energy',
        'loudness',  'speechiness',
       'instrumentalness', 'liveness',  'tempo']].mean()
sorted_year_d = year_d.sort_values(by=['danceability', 'energy','loudness',  'speechiness',
       'instrumentalness', 'liveness',  'tempo'])

high_year_d=sorted_year_d.tail(10)

sorted_year_df= high_year_d.reset_index()

# Extracting relevant columns for plotting
features = ['loudness', 'danceability', 'energy', 'speechiness', 'instrumentalness', 'liveness', 'tempo']
feature_colors = plt.cm.viridis(np.linspace(0, 1, len(features)))

# Create a stacked bar chart
plt.figure(figsize=(12, 8))
bottoms = np.zeros(len(sorted_year_df))

for i, feature in enumerate(features):
    plt.bar(sorted_year_df['year'], sorted_year_df[feature], label=feature, color=feature_colors[i], bottom=bottoms)
    bottoms += sorted_year_df[feature]

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Mean Value')
plt.title('Audio Feature Trends Over the Years')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), title='Audio Features')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()


```
